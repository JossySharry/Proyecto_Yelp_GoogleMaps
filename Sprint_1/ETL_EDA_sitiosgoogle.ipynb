#### Unificamos datasets en un csv 

```python
import pandas as pd
import glob

# Ruta donde están almacenados los archivos JSON Lines
ruta_archivos = r'C:\Users\Alberto\Desktop\Mari\PF\Google Maps_metadata-sitios\metadata-sitios/*.json'

# Lista para almacenar los DataFrames
dataframes = []

# Cargar cada archivo JSON Lines en un DataFrame
for archivo in glob.glob(ruta_archivos):
    df = pd.read_json(archivo, lines=True)  # Leer el archivo JSON Lines
    dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_unificado = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame unificado en un archivo CSV
df_unificado.to_csv('dataset_unificado.csv', index=False)

print("Archivos JSON Lines unificados y guardados como 'dataset_unificado.csv'")

```

#### Limpieza y normalización de columnas:

```python
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)


# Cargar el dataset unificado
df = pd.read_csv(r'C:\Users\Alberto\Desktop\Mari\PF\Google Maps_metadata-sitios\dataset_unificado.csv')

# Verificar duplicados
duplicados = df.duplicated().sum()
print(f"Duplicados encontrados: {duplicados}")

# Eliminar duplicados
df = df.drop_duplicates()
print("Duplicados eliminados.")

# Verificar nulos por columna
nulos = df.isnull().sum()
print("Nulos por columna:")
print(nulos)

# Eliminar filas con valores nulos en columnas críticas
columnas_críticas = ['name', 'category']
df = df.dropna(subset=columnas_críticas)
print("Filas con nulos en columnas críticas eliminadas.")

# Quitar espacios extra o invisibles en nombres de columnas
df.columns = df.columns.str.strip()

# Quitar espacios extra en los valores de tipo texto
df = df.apply(lambda col: col.map(lambda x: x.strip() if isinstance(x, str) else x))

# Verificar tipos de datos
print("Tipos de datos:")
print(df.dtypes)

# Convertir la columna 'category' a minúsculas
df['category'] = df['category'].str.lower()


# Verificar valores únicos en la columna 'category'
print("Categorías únicas:")
print(df['category'].unique())
```

Se identifican columnas con alto porcentaje de nulos por lo que procedemos a eliminarlas: 'address', 'description', 'price', 'hours', 'MISC'.  
También se eliminan filas en las columnas que son relevantes para nuestro análisis: 'name' y 'category'

```python
df = df.drop(columns=['address', 'description', 'price', 'hours', 'MISC'])
df.info() #Confirmamos la eliminación
```

Procedemos a eliminar también las columnas que no vamos a necesitar para nuestro análisis: 'relative_results' ,'state', 'url', 'latitude' y 'longitude'

```python
df = df.drop(columns=['relative_results' ,'state', 'url', 'latitude' , 'longitude'])
df.info() #Confirmamos la eliminación
```

```python
# Verificamos nulos luego de la limpieza
nulos = df.isnull().sum()
print("Nulos por columna:")
print(nulos)
```

#### Análisis de distribución de categorías

```python
# Diccionario de palabras clave asociadas a cada tipo de restaurante latino
categoria_restaurante = {
    'restaurante argentino': ['argentine', 'argentinian', 'argentinian restaurant'],
    'restaurante mexicano': ['mexican', 'mexican restaurant', 'burrito', 'taco', 'pozole', 'Oaxacan'],
    'restaurante colombiano': ['colombian', 'colombian restaurant'],
    'restaurante peruano': ['peruvian', 'peruvian restaurant'],
    'restaurante brasileño': ['brazilian', 'brazilian restaurant'],
    'restaurante cubano': ['cuban', 'cuban restaurant'],
    'restaurante venezolano': ['venezuelan', 'venezuelan restaurant'],
    'restaurante chileno': ['chilean', 'chilean restaurant'],
    'restaurante ecuatoriano': ['ecuadorian', 'ecuadorian restaurant'],
    'restaurante dominicano': ['dominican', 'dominican restaurant'],
    'restaurante centroamericano': ['honduran', 'guatemalan', 'nicaraguan', 'panamanian', 'salvadoran', 'costarican', 'costa rican restaurant', 'honduran restaurant'],
    'restaurante caribeño': ['caribbean', 'caribbean restaurant'],
    'restaurante latinoamericano': ['latin american food', 'latin american', 'south american', 'latin', 'hispanic', 'south american restaurant'],
    'restaurante boliviano': ['bolivian', 'bolivian restaurant'],
    'restaurante paraguayo': ['paraguayan', 'paraguayan restaurant'],
    'restaurante uruguayo': ['uruguayan', 'uruguayan restaurant'],
    'restaurante puertorriqueño': ['puerto rican', 'puerto rican restaurant']
}

# Función para categorizar los restaurantes
def categorizar_restaurante(categoria):
    for categoria_nombre, keywords in categoria_restaurante.items():
        if any(keyword.lower() in categoria.lower() for keyword in keywords):
            return categoria_nombre
    return 'otros'  # Si no coincide con ninguna categoría, se clasifica como 'otros'

# Aplicar la función a la columna 'category' para crear una nueva columna 'restaurante_categoria'
df['restaurante_categoria'] = df['category'].apply(categorizar_restaurante)

# Verificar las primeras filas del DataFrame con la nueva columna 'restaurante_categoria'
print(df[['name', 'category', 'restaurante_categoria']].head())
```

#### Distribución de restaurantes latinoamericanos por tipo

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)


# Contar las categorías de restaurantes
categoria_counts = df['restaurante_categoria'].value_counts()

# Normalizar los valores a un porcentaje del total
categoria_percentage = categoria_counts / categoria_counts.sum() * 100

# Crear el gráfico
plt.figure(figsize=(10, 8))
sns.barplot(x=categoria_percentage.values, y=categoria_percentage.index, palette='viridis', hue=categoria_percentage.index, dodge=False)

# Etiquetas y título
plt.xlabel('Porcentaje de Restaurantes')
plt.ylabel('Categorías')
plt.title('Distribución de Restaurantes Latinoamericanos por Tipo')

# Mostrar el gráfico
plt.show()
```

#### Distribución de calificaciones promedio por tipo de restaurante

```python
# Calcular el promedio de calificaciones para cada tipo de restaurante
categoria_avg_rating = df.groupby('restaurante_categoria')['avg_rating'].mean().sort_values(ascending=False)

# Crear gráfico de barras
plt.figure(figsize=(10, 8))
sns.barplot(x=categoria_avg_rating.values, y=categoria_avg_rating.index, palette='viridis')

# Etiquetas y título
plt.xlabel('Promedio de Calificación')
plt.ylabel('Tipo de Restaurante')
plt.title('Promedio de Calificación por Tipo de Restaurante')

# Mostrar el gráfico
plt.show()

```

Si bien la mayoría de los restaurantes latinoamericanos analizados obtienen calificaciones superiores a 4 estrellas, lo que indica un alto nivel de satisfacción general, es posible identificar algunas diferencias sutiles entre los distintos tipos de cocina. Por ejemplo, los restaurantes uruguayos y venezolanos tienden a obtener calificaciones ligeramente más altas que los restaurantes ecuatorianos y centroamericanos. Estos pequeños matices podrían estar influenciados por factores como la autenticidad de la comida, la experiencia del cliente o la ubicación del restaurante.
